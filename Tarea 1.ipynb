### Partícula en un pozo infinito con fondo irregular

```julia
#Masa de la partícula
m = 1
#longitud de la pared
a=1
λ = [0,0.1,0.5,1];
```

```julia
#Graficando la pared 
x = collect(0.00001:0.001:a-0.00001)
for i in λ
    y= i*cos.(π*x/a)
    plot!(x,y,label="λ=$(i)")
end
#y= λ*cos.(π*x/a)


plot!(ylim = (-1,3),xlabel="x [m]",ylabel="V [J]")
#plot()
```
